## Examples of objects of classes Cube, Parallelepiped, RegularTriangularPrism, WCCoPrism

```python
from randsec.funcs import Cube, Parallelepiped, RegularTriangularPrism, WCCoPrism
```

```python
#Create a Cube object with a side length of 1:
cube = Cube(side=1)

#Find the intersection points of the cube with a plane:
plane = Plane(Point(0.5, 0.5, 0), Vector(1, 0, 0))
intersection_points = cube.intersection_with_plane(plane)
intersection_points
```

```python
#Create a RegularTriangularPrism object with a side length of 1:
prism = RegularTriangularPrism(side=1)

#Find the intersection points of the prism with a plane:
plane = Plane(Point(0.5, 0.5, 0), Vector(1, 0, 0))
intersection_points = prism.intersection_with_plane(plane)
intersection_points
```

```python
#Get coordinates of vertices of prism
prism.iterable_points
```

```python
#Create a Parallelepiped object with side lengths 1, 2, and 3:
parallelepiped = Parallelepiped(side_a=1, side_b=2, side_c=3)

#Find the intersection points of the parallelepiped with a plane:
plane = Plane(Point(0.5, 0.5, 0), Vector(1, 0, 0))
intersection_points = parallelepiped.intersection_with_plane(plane)
intersection_points
```

```python
#Create a WCCoPrism object with specified dimensions:
prism = WCCoPrism(r=1., k=2.)

# Find the intersection points of the prism with a plane:
plane = Plane(Point(0.4, 0.4, 0), Vector(1, 0, 0))
intersection_points = prism.intersection_with_plane(plane)
print('intersection polygon',  intersection_points)

# Find the intersection points of the plane with the prisms basal planes:
intersection_points, intersection_count = prism.intersection_with_basal(plane)
print('intersection polygon',  intersection_points)
print('number of intersections with basal plane', intersection_count)
```

## Geometric characteristics of polygons

```python
from randsec.funcs import get_angles, area_of_polygon_3d, perimeter_3D, semiaxes_from_polygon_3D
from randsec.funcs import minmax_feret, characteristic_of_section

prism = WCCoPrism(r=1., k=2.)
plane = make_random_plane_into_wcco(r=1., k=2.)
secant_polygon = prism.intersection_with_plane(plane)

# Calculate characteristics of the section
angles, area, perim, a_semi, b_semi, min_feret, max_feret, aspect_ratio, sphericity = characteristic_of_section(secant_polygon)

print('Polygon is', secant_polygon)

# Print the results
print("Angles:", angles)
print("Area:", area)
print("Perimeter:", perim)
print("Semi-major axis:", a_semi)
print("Semi-minor axis:", b_semi)
print("Min Feret diameter:", min_feret)
print("Max Feret diameter:", max_feret)
print("Aspect ratio:", aspect_ratio)
print("Sphericity:", sphericity)
```

```python
#Other way to calculate characteristic is usage of funcion:
print("Angles:", get_angles(secant_polygon))
print("Area:", area_of_polygon_3d(secant_polygon))
print("Perimeter:", perimeter_3D(secant_polygon))
print("Semi-major and semi_minor axes: {}".format(semiaxes_from_polygon_3D(secant_polygon)))
```

## Distribution densties

#### Using functions above

```python
from randsec.funcs import Parallelepiped, secant_polygon, area_of_polygon_3d, make_random_plane_into_tetra
from randsec.funcs import norm_dict, extract_and_trim_dict
from matplotlib import pyplot as plt
```

```python
%%time
#For example, to calculate distribution density of area in sections of polyhedra, we need to generate N>>1 sections 
#and store area in sections

areas_dict = {} #empty dict with keys from 0.00 to 10.00 with step 0.01. Need it to store areas in sections
step = 0.01
current_key = 0.00
while current_key <= 10.:
    areas_dict.setdefault(round(current_key, 2), 0)
    current_key += step 

#Parameters of polyhedron to cut
params = {'side_a': 1.,
        'side_b': 2.,
        'side_c': 3.}
#Polyhedron to cut
polyhedron = Parallelepiped(**params)
#Function to generate random sections
way = make_random_plane_into_tetra 
N = 10000 #Do N sections
num_errors = 0 #count errors

#Loop where we count areas in sections
for i in range(N):
    sec_polygon = secant_polygon(polyhedron, way, params=params)
    if len(sec_polygon) > 2:
        area = area_of_polygon_3d(sec_polygon)
        areas_dict[round(area, 2)] += 1
    else:
        num_errors += 1

print('precision is ', 1-num_errors/N)

#To get density distribution we need to normalize histogram we get, so we use norm_dict
#extract_and_trim_dict help us to trim a dictionary to remove trailing zeros and extract keys and values to visualize it

areas_dict = norm_dict(areas_dict, step=0.01)
plt.plot(*extract_and_trim_dict(areas_dict))

plt.title('Density distribution of areas of sections of parallelepiped with a=1, b=2, c=3')
plt.xlabel('area')
plt.ylabel('p')
plt.show()

# if you need to save/load dict you can use:

# np.save(f'areas.npy', areas_dict)
# areas_dict = np.load(f'areas.npy', allow_pickle=True).item()
```

#### From the box

In the "densities" folder of this repository, you can find a variety of densities that were calculated using the HSE supercomputer "CHARISMa".

If you want to download these pre-calculated densities, you should use the PATH (function argument) that contains the "densities" folder. That is, if your path to the folder with the densities is as follows: C:\project\densities, then you should pass the variable PATH='C:\project'.

```python
#There is a function in library, that allows to calculate all distributions from the box.
from randsec.funcs import Parallelepiped, make_random_plane_into_tetra
from randsec.funcs import calculate_all_distributions
import os

#Parameters of polyhedron to cut
params = {'side_a': 1.,
        'side_b': 2.,
        'side_c': 3.}
#Polyhedron to cut
polyhedron = Parallelepiped(**params)

angles, areas, perimeter, a_semi, b_semi, minferet, maxferet, aspect, sphericity = calculate_all_distributions(
    polyhedron=polyhedron,
    way = make_random_plane_into_tetra,
    PATH = os.getcwd(), #in this path in folder densities distibutions will store
    type_figure='parallelepiped',
    n = 1000000,
    params=params
)
```

#### Visualization of densities

```python
from randsec.funcs import extract_and_trim_dict
from matplotlib import pyplot as plt

plt.plot(*extract_and_trim_dict(areas))

plt.title('Density distribution of areas of sections of parallelepiped with a=1, b=2, c=3')
plt.xlabel('area')
plt.ylabel('p')
plt.show()
```

```python
plt.plot(*extract_and_trim_dict(minferet), label='Min Feret diameter')
plt.plot(*extract_and_trim_dict(maxferet), label='Max Feret diameter')

plt.title('Density distribution of feret diameter of sections of parallelepiped with a=1, b=2, c=3')
plt.xlabel('Feret diameter')
plt.ylabel('p')
plt.show()
```

#### Rotate 3D points in the same plane to 2D points in Oxy and back

```python
from randsec.funcs import rotate_points_to_Oxy, rotate_points_back_to_3D
from randsec.funcs import from_list_points_to_array

# Define some 3D points in the same plane
prism = WCCoPrism(r=1., k=2.)
plane = make_random_plane_into_wcco(r=1., k=2.)
points_3d = prism.intersection_with_plane(plane)
points_3d = from_list_points_to_array(points_3d)

print("Some points in 3D:")
print(points_3d)

# Rotate the points to 2D plane (Oxy)
rotated_points_2d, rotation_matrix, z_coord = rotate_points_to_Oxy(points_3d)

print("\nRotated 2D points (Oxy):")
print(rotated_points_2d)
print("Rotation Matrix:")
print(rotation_matrix)
print("Z-coordinate after rotation:", z_coord)

# Rotate the points back to 3D
points_3d_restored = rotate_points_back_to_3D(rotated_points_2d, rotation_matrix, z_coord)

print("\nRestored 3D points:")
print(points_3d_restored)
```
