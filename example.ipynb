##### Draw a min volume ellipse bounding set of point P

```python
import numpy as np
from matplotlib import pyplot as plt

from randsec.funcs import getMinVolEllipse
from randsec.vis import draw_ellipse

P = np.array([[0., 0.],
     [1., 0.],
     [0., 1.]])

center, radii, rotation = getMinVolEllipse(P)

draw_ellipse(center, radii, rotation)
plt.plot(P[:, 0], P[:, 1], 'bo') #points of P
plt.show()
```
