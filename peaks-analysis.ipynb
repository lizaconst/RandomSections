```python
dict_all_peaks = get_dict_all_peaks(PATH,
                                    r_list = [0.2, 0.3, 0.5, 0.6, 0.7, 1],
                                    k_list = [0.6, 0.75, 1, 1.5],
                                    n = 100000)

plt.figure(figsize=(11, 7))

#первый пик, 90 градусов
for k in [0.6, 0.75, 1, 1.5]:
    vals_r = []
    for r in [0.2, 0.3, 0.5, 0.6, 0.7, 1]:
        path = get_path_from_r_k(r, k, n=100000)
        vals_r.append(dict_all_peaks[path]['fm'])
    plt.plot([0.2, 0.3, 0.5, 0.6, 0.7, 1], vals_r, label="k="+str(k), marker='s')
    
    
#второй пик, 120 градусов
for k in [0.6, 0.75, 1, 1.5]:
    vals_r = []
    for r in [0.2, 0.3, 0.5, 0.6, 0.7, 1]:
        path = get_path_from_r_k(r, k, n=100000)
        vals_r.append(dict_all_peaks[path]['sm'])
    plt.plot([0.2, 0.3, 0.5, 0.6, 0.7, 1], vals_r, label="k="+str(k), marker='H')

plt.xlabel('r')
plt.ylabel('peak')
plt.legend(loc='upper right')
plt.title('Положение пиков плотностей распределений при различных r, k')
plt.savefig(PATH + "peaks.pdf", format='pdf')
```
